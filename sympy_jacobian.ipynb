```python
import sympy as smp
smp.init_printing()
```

```python
N, R0, D, S, I, P, t = smp.symbols('N R0 D S I P t')
```

```python
P = smp.Matrix([
    R0, 
    D,
    S - R0 * S * I * t / (N * D),
    S + R0 * S * I * t / (N * D) - I * t / D,
])
P
```

```python
J = P.jacobian([R0, D, S, I])
J
```

```python
s_R0R0, s_R0D, s_R0S, s_R0I = smp.symbols('sigma_R0R0  sigma_R0D sigma_R0S sigma_R0I')
s_DD, s_DS, s_DI = smp.symbols('sigma_DD, sigma_DS, sigma_DI')
s_SS, s_SI = smp.symbols('sigma_SS, sigma_SI')
s_II = smp.symbols('sigma_II')

s = smp.Matrix([
    [s_R0R0, 0,    0,    0],
    [0,      s_DD, 0,    0],
    [0,      0,    s_SS, 0],
    [0,      0,    0,    s_II],
])
s
```

```python
s2 = J.T @ s @ J
s2
```
