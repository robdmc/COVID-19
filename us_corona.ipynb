# Load Data

```python
class Loader:
    file_name = './csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
    
    def __init__(self, start_date_str='3/1/2020'):
        self.start_date_str = start_date_str

    @ezr.cached_container
    def df_raw(self):
        
        df = pd.read_csv(self.file_name)
        df = df[df['Country/Region'] == 'US']
        df = df.rename(columns={'Province/State': 'location', 'Country/Region': 'country', 'Lat': 'lat', 'Long': 'lon'})
        return df

    @ezr.cached_container
    def df(self):
        df = self.df_raw
        df = df[~df.location.str.contains('Princess')].sort_values(by='location')
        df = df.set_index(['country', 'location', 'lat', 'lon']).T
        df.index = pd.DatetimeIndex(df.index.values)
        df['cases'] = df.sum(axis=1)
        return df
    
    @ezr.cached_container
    def df_total(self):
        df = self.df
        df = df[['cases']].reset_index()
        df.columns=['date', 'cases']
        df['days'] = (df.date - pd.Timestamp(self.start_date_str)).dt.days
        df = df[df.days >= 0]
        return df
    
    def plot_map(self):
        df = loader.df.T
        df['total'] = df.sum(axis=1)
        df = df[df.total > 0]
        df = df[['total']].reset_index()
        df = df[df.total > 0]
        df.loc[:, 'lat'] = [np.NaN if l == '' else float(l) for l in df.lat]
        df.loc[:, 'lon'] = [np.NaN if l == '' else float(l) for l in df.lon]
        df = df[df.lat.notnull() & df.lon.notnull()]

        m = folium.Map(location=[35.29, -97.766], zoom_start=4, left='10%', width='80%', height='70%')

        for tup in df.itertuples():
            radius =  3 * np.log10(tup.total)
            folium.CircleMarker([tup.lat, tup.lon], radius=radius, tooltip=f'{tup.location} {tup.total}', popup=f'<b>{tup.location}<br>{tup.total} cases</b>').add_to(m)
        display(m)
    

loader = Loader()
loader.plot_map()
```

```python
class ModelBase:
    FUTURE_DAYS = 7
    CUSTOM_COST = True
    
    def __init__(self, name, start_date_str='3/1/2020',  **params):
        self.name = name
        self.start_date_str = start_date_str
        self.future_days = self.FUTURE_DAYS
        
        self.loader = Loader(start_date_str)
        df = self.loader.df_total
        
        self.df = df.copy()
        self.fitter = ezr.Fitter(**params)
        self._fit()
        self.future_days = self.FUTURE_DAYS
        
    def model(self, p):
        raise NotImplementedError('You must define this')
        
    def cost(self, p):
        y_obs = p.y
        sigma = np.sqrt(np.maximum(y_obs, 0))
        y_fit = self.fitter._model(p)
        
        err = (y_fit - y_obs) / sigma
        return np.sum(err ** 2)        
        
    def _fit(self):
        df = self.df
        if self.CUSTOM_COST:
            cost = self.cost
        else:
            cost = None
        self.fitter.fit(x=self.xd, y=self.yd, model=self.model, cost=cost, plot_every=10, algorithm='fmin_powell')

    @ezr.cached_container
    def xd(self):
        return self.df.days.values
    
    @ezr.cached_container
    def yd(self):
        return self.df.cases.values

    @ezr.cached_container
    def xf(self):
        df = self.df
        return np.arange(df.days.max() + self.future_days + 1).astype(np.float64)
    
    @ezr.cached_container
    def yf(self):
        return self.fitter.predict(self.xf)
    
    @property
    def title(self):
        return '  '.join([self.name] + [f'{k}={v:0.1f}' for k, v in self.fitter.params.as_dict().items()])
    
    def plot_hv(self):
        epoch = pd.Timestamp(self.start_date_str).to_pydatetime()
        date_d = [epoch + datetime.timedelta(days=int(d)) for d in self.xd]
        date_f = [epoch + datetime.timedelta(days=int(d)) for d in self.xf]
        
        c1 = hv.Scatter((date_d, self.yd), label='Confirmed').options(
            size=5, logy=False, color=ezr.cc.a, show_grid=True
        ).redim(x=f'Days since {self.start_date_str}', y='Cases')
        
        c2 = hv.Curve((date_f, self.yf), 'Days', 'Cases (K)', label=self.title).options(color=ezr.cc.b)
        
        c3 = c1.options(logy=True)
        c4 = hv.Scatter(c2).options(color=ezr.cc.b, size=5)
        display(c4 * c1 * c2)
        display(c4 * c3 * c2)
    
#     def plot_hv(self):
#         c1 = hv.Scatter((self.xd, self.yd), label='Confirmed').options(
#             size=5, logy=False, color=ezr.cc.a, show_grid=True
#         ).redim(x=f'Days since {self.start_date_str}', y='Cases')
#         c2 = hv.Curve((self.xf, self.yf), 'Days', 'Cases', label=self.title).options(color=ezr.cc.b)
        
#         c3 = c1.options(logy=True)
#         display((c1 * c2 + c3 * c2).cols(1))
        
    def plot_mpl(self, ax=None, color=ezr.cc.b):
        if ax is None:
            ax = ezr.figure()
        ax.scatter(self.xd, self.yd, color=ezr.cc.a)
        ax.plot(self.xf, self.yf, color, label=self.title)
        ax.set_xlabel(f'Days since {self.start_date_str}')
        ax.set_ylabel('Cases')
        ax.legend(loc='best')
        return ax
        
    def plot(self):
        display(Markdown(f"# {self.name} Model"))
        self.plot_hv()
        self.plot_mpl()
        
        
```

```python
class ModelExponent(ModelBase):
    def model(self, p):
        return  np.exp((p.x - p.t0) / p.tau)        


model_exp = ModelExponent(name='Exponential', t0=5,  tau=3)
model_exp.plot()
```

```python
class ModelLogistic(ModelBase):
    def model(self, p):
        z = (p.x - p.t0) / p.tau
        return np.exp(p.a) * np.exp(z) / (1 + np.exp(z))

    
model_log = ModelLogistic(name='Logistic', a=12, tau=3, t0=10)
model_log.plot()
```
