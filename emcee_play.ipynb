# Load Data

```python
class Loader:
    glob_str = './csse_covid_19_data/csse_covid_19_daily_reports/*.csv'
    
    COUNTRY = 'US'
#     COUNTRY = 'UK'
#     COUNTRY = 'Italy'
#     COUNTRY = 'Germany'
#     COUNTRY = 'Hong Kong'

    def __init__(self, start_date_str='3/18/2020', raw=False):
        print(f'*** Estimating for {self.COUNTRY} ***')
        self.start_date_str = start_date_str
        self.start_date = parse(start_date_str)
        self.raw = raw
        
    @ezr.cached_container
    def df(self):
    
        files = sorted(glob.glob(self.glob_str))
        date_strings = [os.path.basename(f).replace('.csv', '') for f in files]
        dates = [parse(d) for d in date_strings]

        files_dates = sorted(zip(files, dates))
        files_dates = [t for t in files_dates if t[1] >= self.start_date]
        

        df_list = []
        for file, date in files_dates:
            df = pd.read_csv(file)
            df.columns = ezr.slugify(df.columns)
            if self.raw:
                return df
            df = df[df['country_region'] == self.COUNTRY]
            df['date'] = date
            df = df[[
                'date',
                'province_state',
                'confirmed',
                'deaths',
                'recovered'
            ]].fillna(0)
            df_list.append(df)

        df = pd.concat(df_list, ignore_index=True, sort=False)
        df = df.groupby(by='date')[['confirmed', 'deaths', 'recovered']].sum().reset_index()
        
        df['N'] = 331e6
        df['infected'] = df.confirmed - df.recovered - df.deaths
        df['days'] = (df.date - df.date.min()).dt.days
        
        
        return df
    
    def plot(self):
        c1 = hv.Curve((self.df.date, self.df.confirmed + 1), label='Confirmed').options(logy=True)
        c2 = hv.Curve((self.df.date, self.df.recovered + self.df.deaths + 1), label='Recovered/Died')
        c3 = hv.Curve((self.df.date, self.df.infected + 1), label='Infected')
        display(c1 * c2 * c3)

loader = Loader(raw=False)
loader.plot()
loader.df.tail()
```

```python
def deriv(p, y):
    R0 = np.exp(p.lR0)
    D = np.exp(p.lD)
    nu = 1. / D
    N = p.N
    beta = R0 / (N * D)

    C, I = y[0], y[1]
    C_dot = beta * (N - C) * I
    I_dot = beta * (N - C) * I - nu * I
    
    return np.array([C_dot, I_dot])


def model(p):
    t = p.t
    I0, C0 = np.exp(p.lI0), np.exp(p.lC0)
    y0 = np.array([C0, I0])
    
    dd = lambda t, y: deriv(p, y)

    res = solve_ivp(dd, (t[0], t[-1]), y0, t_eval=t)
    return res.y



def plotter(data):
    t, y, yf, r0, d = data
    
    c_d = y[0, :]
    i_d = y[1, :]
    
    c_f = yf[0, :]
    i_f = yf[1, :]
    
    c_c_d = hv.Scatter((t, c_d), 'days', 'Cumulative').options(size=5, logy=True)
    c_c_f = hv.Curve((t, c_f), label=f'R0={r0:0.2f}   D={d:0.2f}').options(color=ezr.cc.b, logy=True)
    
    c_i_d = hv.Scatter((t, i_d), 'days', 'Infected').options(size=5, logy=True)
    c_i_f = hv.Curve((t, i_f), label=f'R0={r0:0.2f}   D={d:0.2f}').options(color=ezr.cc.b)
    
    return hv.Layout([
        c_c_d * c_c_f,
        c_i_d * c_i_f
    ]).cols(1)

anim = ezr.Animator(plotter, plot_every=2)

def log_prob(a, p, pipe):
    
    # Injest the current paramaters
    p.ingest(a)
    
    # Run the model to find the fit
    yf = model(p)

    # Fractional errors at about 25% (1 dB)
    z = 10 * np.log10(yf/p.y) / 1
    z2 = z ** 2
    
    # Log prob of the data with the current params
    lp = -np.sum(z2)
    
    r0, d, c0, i0 = np.exp(p.array)
#     anim.send((p.t, p.y, yf, r0, d))
    
    # Include parameter priors
    lp -= ((r0 - p.pri_R0) / p.sig_R0) ** 2
    lp -= ((d - p.pri_D0) / p.sig_D) ** 2
    lp -= ((c0 - p.pri_C0) / p.sig_C0) ** 2
    lp -= ((i0 - p.pri_I0) / p.sig_I0) ** 2
    
    return lp
    

def log_error(a, p, pipe):
    # Injest the current paramaters
    p.ingest(a)
    
    # Run the model to find the fit
    yf = model(p)
    
    r0, d, c0, i0 = np.exp(p.array)
    anim.send((p.t, p.y, yf, r0, d))
    return -log_prob(a, p, pipe)
    
    

p = ezr.ParamState(
    lR0=np.log(3.), 
    lD=np.log(20.), 
    lC0=np.log(200.), 
    lI0=np.log(200.)
)

df = loader.df
p.given(
    N=331e6, # Pop of US,
    
    t=df.days.values,
    y=df[['confirmed', 'infected']].T.values,
    
    pri_R0=2.5,
    pri_D0=17.,
    pri_C0=200.,
    pri_I0=200.,
    
    sig_R0=3.5,
    sig_D=17.,
    sig_C0=200.,
    sig_I0=200.
)

pipe = Pipe(data=[])

res = fmin_powell(log_error, p.array, args=(p, pipe))
plog = p.clone()
plog.ingest(res)
p.ingest(np.exp(res))
plin = p.clone()
p.drop('t', 'y')
p
```

```python
n_walkers = 200
n_dim = len(plog.array)
pos = plog.array + 1e-4 * np.random.randn(n_walkers, n_dim)

with Pool(processes=8) as pool:
    sampler = emcee.EnsembleSampler(nwalkers=n_walkers, ndim=n_dim, log_prob_fn=log_prob, args=(plog, pipe), pool=pool)
    state = sampler.run_mcmc(pos, 1000, progress=True, )
```

```python
def get_curve(a, ppred):
    today = loader.start_date
    t = [today + datetime.timedelta(days=int(d)) for d in ppred.t]
    y = model(ppred)
    c = hv.Curve((t, y[1, :]), 'Date', 'Infected').options(color=ezr.cc.a)
    return c


ppred = plog.clone()
ppred.drop('y')
ppred.t = np.arange(180)

today = loader.start_date
d = [today + datetime.timedelta(days=int(d)) for d in plog.t]
c_list = []
n_curves = 100
for tup in dfs_log.sample(n_curves).iterrows():
    a = tup[1].values
    ppred.ingest(a)
    c = get_curve(a, ppred).options(alpha=.05, logy=True)
    c_list.append(c)
    
ppred_map = plog.clone()
ppred_map.t = ppred.t
ppred_map.y = model(ppred_map)
dp = [today + datetime.timedelta(days=int(d)) for d in ppred_map.t]

c_list.append(hv.Curve((dp, ppred_map.y[1, :])).options(alpha=1, color=ezr.cc.b, line_width=2))
c_list.append(hv.Scatter((d, plog.y[1,:])).options(color=ezr.cc.c, size=5) )
ol = hv.Overlay(c_list)
ol
```
